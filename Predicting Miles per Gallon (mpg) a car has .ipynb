## Training the model

```r
# fittimg the linear regression on the data
model<-lm(mpg~cyl+disp+hp+drat+wt+qsec+vs+am+gear+carb, data=cars)
```

## Evaluating the model performance

```r
summary(model)
```

Number of cylinders have a negative relation to mpg. An increase in the number of cyl by 1 reduces mpg by 0.1  
Increase in displacement by an inch increases mpg by 0.11  
Increase in rear axle ratio by one increases mpg by 0.78  
Increase in weight by a pound reduces the mpg by 3.71  
  
All estimates are not statistically significant apart from wt which is significant at 5 $\%$ significant level. However, the $r^2$ shows that 86.9 $\%$ of the variation in explanatory variable explain the variation in mpg variable.  

## Improving the model performance

```r
# adding non-linear parameters to the disp and hp
cars$disp2<-cars$disp^2
cars$hp2<-cars$hp^2
head(cars)
```

```r
model1<-lm(mpg~cyl+disp+hp+drat+wt+qsec+vs+am+gear+carb+disp2+hp2, data=cars)
summary(model1)
```

$r^2$ has also improved

```r
# doing a binanry transformation on cyl variable
cars$cyl6<-ifelse(cars$cyl < 6, 0,1)
head(cars)
```

The reason to choose cylinder is because the predictor graph against mpg shows three lines at 4,6 and 8. Most values at the middle lie below the avaerage of 20 on the response variable hence are made be at the same binanry level as 8.

```r
model2<-lm(mpg~cyl+disp+hp+drat+wt+qsec+vs+am+gear+carb+disp2+hp2+cyl6, data=cars)
summary(model2)
```

The $R^2$ has improved to 91.05$\%$

```r
plot(cars$hp, cars$wt)
```

```r
# interaction effect with wt and hp
# They both move in the same direction hence similar effect on the response variable
model3<-lm(mpg~wt*hp, data=cars)
summary(model3)
```

```r
# The new model taking into account the changes
model4<-lm(mpg~cyl+disp+drat+qsec+vs+am+gear+carb+disp2+hp2+cyl6+wt*hp, data=cars)
summary(model4)
```

The model improves with $R^2$ of 91.62$\%$.  
However, most variables are not statistically significant when use all together within the model.
