```r
# importing the dataset DATSET.csv
data<-read.csv("DATASET.csv")
str(data)
```

```r
# converting categorical variables to factors
data$is_above_mean_price<-factor(data$is_above_mean_price)
data$waterfront<-factor(data$waterfront)
data$view<-factor(data$view)
data$condition<-factor(data$condition)

data$id<- NULL # removing the id variable
str(data)
```

```r
# creating normalizing function 
normalize<-function(x){
  return ((x-min(x))/(max(x)-min(x)))
}
```

```r
# Testing the function
y<-c(1,2,4)
normalize(y)
```

```r
# Applying the noramlization to the dataset
# splitting the data into factors and numerics
num<-data[,c(-2,-8,-9,-10)]
fac<-data[,c(2,8,9,10)]
str(num)
str(fac)
```

```r
# normalizing the numeric data
num_n<-lapply(num[1:16], normalize)
str(num_n)
```
