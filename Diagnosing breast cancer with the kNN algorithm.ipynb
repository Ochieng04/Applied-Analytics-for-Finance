```r
# importing Breast Cancer Wisconsin Diagnostic dataset as wc
wc<- read.csv("wisc_bc_data.csv", stringsAsFactors=F)
```

```r
head(wc)
```

```r
str(wc)
```

```r
#Removing the first column "ID" as it would affect our prediction
wc<-wc[-1]
head(wc)
```

```r
#Checking the values of diagnosis
table(wc$diagnosis)
```

```r
prop.table(wc$diagnosis)
```

```r
#We therefore need to record the diagnosis values as a factors
wc$diagnosis<-factor(wc$diagnosis, levels = c("B","M"), labels = c("Benign","Malignant"))
head(wc)
```

```r
#Checking the proportion of the factors in diagnosis column
round(prop.table(table(wc$diagnosis))*100, digits=1)
```

```r
# summary of three main features
summary(wc[c("radius_mean","area_mean","smoothness_mean")])
# The range for area mean is too large and would have an impact on ditermining the Euclidean distance
```

```r
# Normalizing the three data by creating a normalize function
normalize<-function(x){
    return((x-min(x))/(max(x)-min(x)))
}
```

```r
# We normalize each feature in the dataframe
wc_n<-as.data.frame(lapply(wc[2:31], normalize))
head(wc_n)
```

```r
summary(wc_n[c("radius_mean","area_mean","smoothness_mean")])
```

## Creating Training and Test Data

```r
# Check the number of rows
nrow(wc_n)
```

```r
# Divide the datasets into about 80% for training our model and 20% for testing.
wc_train<-wc_n[1:469,]
wc_test<-wc_n[470:569,]
```

```r
# Creaing class labels for the diagnosis variable
wc_train_labels<-wc[1:469,1]
wc_test_labels<-wc[470:569,1]
```

## Training the model

```r
# Using the knn() function to classify the test data
library(class)
wc_test_pred <- knn(train = wc_train, test = wc_test,cl = wc_train_labels, k=21)
```

## Evaluating model perfomance

```r
# install the package 'gmodels'
# install.packages('gmodels')
```

```r
library(gmodels)
CrossTable(wc_test_labels, wc_test_pred, prop.chisq = FALSE)
# 77 results were True negatives (TN); Benign predicted benign
# 21 results predicted malignant, True positives
# 2 results predicted Benign while it was malignant; False negative (FN)
# NOTE: The values may keep changing as you rerun the whole script
```

```r
# The model is 98% accurate
```
